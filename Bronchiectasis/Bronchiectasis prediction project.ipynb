```python
import pandas as pd
df=pd.read_excel("Bronchiectasis_Prediction_Dataset.xlsx")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_gender = LabelEncoder()
le_sputum_color = LabelEncoder()
le_chronic_cough = LabelEncoder()
le_shortness_of_breath = LabelEncoder()
le_clubbing =LabelEncoder()
le_Previous_Resp_Infections = LabelEncoder()
le_fatigue = LabelEncoder()
```

```python
df['Gender'] = le_gender.fit_transform(df['Gender'])
df['Sputum_Color'] = le_sputum_color.fit_transform(df['Sputum_Color'])
df['Chronic_Cough']=le_chronic_cough.fit_transform(df['Chronic_Cough'])
df['Shortness_of_Breath']= le_shortness_of_breath.fit_transform(df['Shortness_of_Breath'])
df['Clubbing'] = le_clubbing.fit_transform(df['Clubbing'])
df['Previous_Resp_Infections'] =le_Previous_Resp_Infections.fit_transform(df['Previous_Resp_Infections'])
df['Fatigue']=le_fatigue.fit_transform(df['Fatigue'])
```

```python
df.head()
```

```python
df = df.drop(df[['Patient_ID','Age']],axis = 'columns')
```

```python
df.head()
```

```python
target = df['Bronchiectasis (Label)']
```

```python
inputs = df.drop(['Bronchiectasis (Label)'],axis = 'columns')
```

```python
inputs.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import cross_val_score
scores_svm =cross_val_score(SVC(),inputs,target)
scores_dt = cross_val_score(DecisionTreeClassifier(),inputs,target)
scores_rf = cross_val_score(RandomForestClassifier(n_estimators=100),inputs,target)
```

```python
scores_svm
```

```python
scores_dt
```

```python
scores_rf
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
